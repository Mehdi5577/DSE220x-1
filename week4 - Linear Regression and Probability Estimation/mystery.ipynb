```python
def feature_subset_regression(x,y,flist):
    if len(flist) < 1:
        print("Need at least one feature")
        return
    for f in flist:
        if (f < 0) or (f > 100):
            print("Feature index is out of bounds")
            return
    regr = linear_model.LinearRegression()
    regr.fit(x[:,flist], y)
    y_pred = regr.predict(x[:,flist])
    return mean_squared_error(y, y_pred)
```

### Using reduced features to train again
- reduce one feature each time and calculate the MSE

```python
non_zero_index = np.array([ 1,  2,  4,  6,  7, 10, 12, 14, 15, 16, 18, 19, 22, 28, 31, 47, 53,
       56, 68, 72, 81, 98])
```

```python
print(non_zero_index)
print(non_zero_index.shape)
```

```python
loo = LeaveOneOut()
lamb_best = 0.06
def loo_train(non_zero_index, X_train=X_train, X_test=X_test, y_train=y_train, y_test=y_test, alpah=lamb_best):
    loss_all = []  # collect all loss
    inx_all = []  # collect all index
    for train_inx, other_inx in loo.split(non_zero_index):
        inx_all.append(non_zero_index[train_inx])
        X_train_new = X_train[:, non_zero_index[train_inx]]
        X_test_new = X_test[:, non_zero_index[train_inx]]
        lasso_reg = Lasso(alpha=alpah)
        lasso_reg.fit(X_train_new, y_train)
        # print('--------{}---------'.format(i))
        # print(lasso_reg.intercept_, lasso_reg.coef_)
        # print(np.where(lasso_reg.coef_!=0)[0].shape)
        y_pred = lasso_reg.predict(X_test_new)
        loss_all.append(mean_squared_error(y_pred, y_test))
    return loss_all, inx_all
```

```python
loss_all, inx_all = loo_train(non_zero_index)
min_loss = np.min(loss_all)
print(min_loss)
inx_all[np.argmin(loss_all)]
```

```python
def exe_loo_training(non_zero_index, t=10, min_loss=3):
    for i in range(t):
        loss_all, inx_all = loo_train(non_zero_index)
        min_loss_current = np.min(loss_all)
        # print(min_loss_current)
        if min_loss_current < min_loss:
            non_zero_index = inx_all[np.argmin(loss_all)]
            min_loss = min_loss_current
            print(min_loss)
        else:
            return non_zero_index, min_loss
    return non_zero_index, min_loss         
```

```python
min_loss = 0.936050614538
non_zero_index, min_loss = exe_loo_training(non_zero_index, t=10, min_loss=min_loss)
```
